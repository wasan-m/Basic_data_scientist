#Function
* ในการทำงานจริง โค้ดของเราจะประกอบด้วยชุดคำสั่งมากมายที่มีการเรียกใช้ซ้ำบ่อย เพื่อความสะดวกสบายในการเรียกใช้งาน เราสามารถกำหนดชุดคำสั่งเหล่านั้นให้เป็นฟังก์ชั่นได้ เมื่อต้องการใช้งาน เราสามารถเรียกชุดคำสั่งนั้นจากฟังก์ชั่นได้เลย ไม่จำเป็นต้องพิมพ์ชุดคำสั่งใหม่ทั้งหมด

##การสร้างและเรียกใช้งาน function

```
def print_hello():
  print('Hello World')
```

```
print_hello()
```

##Arguments
* ตัวแปรที่ต้องการจะป้อนเป็นเข้าไปในฟังก์ชั่น
* สามารถกำหนด argument ได้หลายตัว แต่ละตัวคั่นด้วยเครื่องหมาย comma (,)

```
#โดยทั่วไป ภายในฟังก์ชั่นจะใส่คำอธิบายไว้ใน triple quote เพื่ออธิบายการทำงานของฟังก์ชั่น
def sub_function(a, b):
  '''
  function for subtracting integer
  '''
  return a-b
```

```
sub_function(1,4)
```

```
result = sub_function(1,4)
result
```

```
#ถ้าค่า input เป็น string จะ error เพราะ string ไม่สามารถลบกันได้
sub_function('hello', 'world')
```

```
#สามารถกำหนดค่า argument ให้เป็น list ได้
def print_list(list):
  for n in list:
    print(n)
```

```
print_list([1,2,3,4,5])
```

## Default Paramter

```
#สามารถกำหนดค่า default ของ argument ได้
def add_function(a, b=5):
  return a+b
```

```
add_function(7)
```

```
add_function(5,6)
```

## Example
* เขียนฟังก์ชั่นเพื่ออ่านค่าจาก list ของตัวเลข หากมีค่ามากกว่า threshold โดยให้ค่า default threshold เท่ากับ 10 หากตัวเลขมากกว่า threshold จะ return 'High' หากมีค่าน้อยกว่าเท่ากับ threshold จะ return 'Low'

```
def check_high_low(list_num, threshold=10):
  for i in list_num:
    if i > threshold:
      print('{} is High'.format(i))
    else:
      print('{} is Low'.format(i))
```

```
check_high_low([0,1,3,6,7,8,9],threshold=5)
```

```
import numpy as np
list_num = np.random.randint(20, size=20)
check_high_low(list_num, threshold=10)
```
