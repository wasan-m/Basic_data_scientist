## Dataframe คืออะไร
* ข้อมูลที่มีลักษณะเป็นตารางขนาด 2 มิติ ประกอบด้วย Row และ Column
 * Row จะเก็บข้อมูลที่เป็น Record เช่น รายชื่อนักเรียนแต่ละคนหรือรายชื่อลูกค้าแต่ละคน
  * Column จะเก็บข้อมูลที่เป็นคุณลักษณะหรือ Feature ของแต่ละ Record
* ใน Pandas สามารถอ่านไฟล์ที่มีลักษณะเป็นตารางเช่นไฟล์ csv, xlsx, table, sql แล้วนำมาเก็บในรูป Dataframe ได้
* นอกจากนั้นยังสามารถสร้าง Dataframe โดยใช้ List หรือ Dictionary ได้เช่นกัน

### สร้าง Dataframe จาก List

```
list = [['a','b','c'],['d','e','f'],['g','h','i']]
df = pd.DataFrame(list)
df
#หากไม่กำหนดชื่อ column จะถูกตั้งชื่อเป็น Index ของ column นั้นๆ
```

##การใช้งาน Duplicated และ drop_duplicated

```
df3 = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 4, 3.5, 15, 5]
})
```

```
df3
```

```
#ใช้คำสั่ง duplicated เพื่อเช็คว่ามีแถวไหนซ้ำกันบ้าง โดยแถวแรกจะขึ้น False ส่วนแถวที่ตามมาที่เหลือจะขึ้น True
df3.duplicated()
```

```
#ในกรณีที่ตั้ง keep=last ค่าสุดท้ายของแถวที่ซ้ำกันจะขึ้น False ส่วนที่เหลือจะเป็น True
#ค่า default คือ keep='first'
df3.duplicated(keep='last')
```

```
#ในกรณีที่ตั้ง keep=False แถวที่มีการซ้ำทั้งหมดจะขึ้น True
df3.duplicated(keep=False)
```

```
#สามารถ check duplicated เฉพาะ column
df3.duplicated(subset=['brand'])
```

```
#ใช้คำสั่ง drop_duplicates เพื่อ drop แถวที่มีค่าซ้ำ
df3.drop_duplicates()
```

```
#ใช้คำสั่ง drop_duplicates เพื่อ drop แถวที่มีค่าซ้ำเฉพาะ columns
df3.drop_duplicates(subset=['brand'])
```

```
#drop แถวที่มีค่าซ้ำโดยอ้างอิงค่าจากแถว brand และ style โดยเก็บค่าสุดท้ายเอาไว้
df.drop_duplicates(subset=['brand', 'style'], keep='last')
```

##การใช้งาน crosstab
* สร้าง cross-tabulatation ระหว่างสองตัวแปรหรือมากกว่า การใช้งานใกล้เคียงกับ pivot table

```
#สร้างตารางโดยให้ index เป็นค่า Stage และ column เป็นค่า Attribute ส่วน value ในตารางเป็นค่าความถี่ในแต่ละ Stage และ Attribute
pd.crosstab(df['Stage'], df['Attribute'])
```

```
#สร้างตารางโดยให้ index เป็นค่า Stage และ column เป็นค่า Attribute และ value เป็นค่า LV 50 HP โดยให้ aggfunc เท่ากับ mean
pd.crosstab(df['Stage'], df['Attribute'], values=df['Lv 50 HP'], aggfunc='mean')
```

##Filter ตามเงื่อนไข
* สามารถกำหนด condition เพื่อ filter dataframe ได้ โดยใช้เครื่องหมาย <, > และ ==
* สามารถกำหนดหลาย condition โดยเชื่อแต่ละ condition ด้วย 'and' กับ 'or'

```
#Filter ข้อมูลโดยกำหนดเงื่อนไขว่าค่า Lv 50 HP ต้องมากกว่า 500
df[df['Lv 50 HP']>500]
```

```
#Filter ข้อมูลโดยกำหนดเงื่อนไขว่าค่า Lv 50 HP ต้องมากกว่า 500 และ Lv50 Atk ต้องน้อยกว่าเท่ากับ 100
df[(df['Lv 50 HP']>500)&(df['Lv50 Atk']<100)]
```

```
#Filter ข้อมูลโดยกำหนดเงื่อนไขว่า Type ต้องเป็น Free หรือ Data และ Lv 50 HP ต้องมากกว่าเท่ากับ 5
df[((df['Type']=='Free')|(df['Type']=='Data'))&(df['Lv 50 HP']>=500)]
```

```
#สามารถใช้ร่วมกับคำสั่ง loc และ iloc
#Filter ข้อมูลโดยกำหนดเงื่อนไขว่า Type ต้องเท่ากับ Free และให้เลือก column Digimon
df.loc[df['Type']=='Free', 'Digimon']
```

##การใช้งาน apply
* สามารถ apply ฟังก์ชั่นที่เรากำหนดเองเข้ากับค่าใน column ได้

```
import numpy as np
```

```
#ตัวอย่างฟังก์ชั่นนับจำนวนตัวอักษร
def count_char(char):
  return len(char)
```

```
#Apply ฟังก์ชั่น count_char ที่คอลัมน์ Digimon ผลลัพธ์ที่ได้คือจำนวนตัวอักษรในชื่อ Digimon
df['Digimon'].apply(count_char)
```

```
#สามารถใช้ Lambda Expression แทนการเขียนฟังก์ชั่น
df['Digimon'].apply(lambda x: len(x))
```

```
#เขียนรูปแบบนี้ได้ ได้ผลลัพธ์เหมือนกัน
df.apply(lambda x: len(x['Digimon']), axis=1)
```
